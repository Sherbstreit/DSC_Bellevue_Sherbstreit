```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import date
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_excel('CANDY-HIERARCHY-2015-SURVEY-Responses.xlsx')
```

```python
# set df display parameters
desired_width = 100
pd.set_option('display.width', desired_width)
np.set_printoptions(linewidth=desired_width)
pd.set_option('display.max_columns', 15)
df.head()
```

```python
# set NaN threshold and remove columns with greater than 30% NaN
limitPer = len(df) * .30
df = df.dropna(thresh=limitPer,axis=1)
```

```python
# clean column headers
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[', '').str.replace(']', '')
```

```python
# look clean at variable names
for col in df.columns: 
    print(col) 
```

```python
# drop irrelevant headers
df.drop(['are_you_going_actually_going_trick_or_treating_yourself?',
         'anonymous_brown_globs_that_come_in_black_and_orange_wrappers',
        'any_full-sized_candy_bar', 'bonkers', 'brach_products_(not_including_candy_corn)',
        'vials_of_pure_high_fructose_corn_syrup,_for_main-lining_into_your_vein',
        'candy_that_is_clearly_just_the_stuff_given_out_for_free_at_restaurants',
        'cash,_or_other_forms_of_legal_tender', 'dental_paraphenalia',
        'generic_brand_acetaminophen', 'glow_sticks', 'broken_glow_stick',
        'gum_from_baseball_cards', 'creepy_religious_comics/chick_tracts',
        'healthy_fruit', 'hugs_(actual_physical_hugs)', 'jolly_rancher_(bad_flavor)',
        'kale_smoothie', 'kinder_happy_hippo', 'fuzzy_peaches', 'lapel_pins', 'licorice_(not_black)',
        'joyjoy_(mit_iodine)', 'reggie_jackson_bar', 'pencils', 'senior_mints',
        'mint_kisses', 'mint_juleps', 'mint_leaves', 'ribbon_candy',
        'chick-o-sticks_(we_don’t_know_what_that_is)', 'spotted_dick',
        'peterson_brand_sidewalk_chalk', 'peanut_butter_jars',
        'vicodin', 'white_bread','whole_wheat_anything',
        'guess_the_number_of_mints_in_my_hand.', 'betty_or_veronica?',
        'check_all_that_apply:_"i_cried_tears_of_sadness_at_the_end_of______________"',
        '"that_dress*_that_went_viral_early_this_year_-_when_i_first_saw_it,_it_was_________"',
        'what_is_your_favourite_font?',
        'if_you_squint_really_hard,_the_words_"intelligent_design"_would_look_like.',
        'fill_in_the_blank:_"imitation_is_a_form_of_____________"',
        'sea-salt_flavored_stuff,_probably_chocolate,_since_this_is_the_"it"_flavor_of_the_year',
        'which_day_do_you_prefer,_friday_or_sunday?'], 1, inplace=True)
```

```python
# drop all columns with the word 'please' in them
df = df[df.columns.drop(list(df.filter(regex='please')))]
```

```python
# fill NaN values with mode
for column in df.columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
# convert timestamp to date only format
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date
df.head()
```

```python
# prepare second df
df2 = df.drop(['timestamp','how_old_are_you?'], axis=1)
```

```python
# count instances of joy and despair
df_count = df2.apply(pd.Series.value_counts)
df_count.head()
```

```python
# transpose df make joy and despair columns
df_count = df_count.transpose()
```

```python
# create net approve row
df_count['Net_Approve'] = df_count['JOY'] - df_count['DESPAIR']
```

```python
# create overall ranking row
df_count['ranking'] = df_count['Net_Approve'] / len(df2)
```

```python
# remove joy and despair
df_count.drop(['DESPAIR', 'JOY'], 1, inplace=True)
df_count = df_count.transpose()
df_count.head()
```

```python
# join the two dataframes
pd.concat([df, df_count], axis=0)
```

```python
# rename age column
df = df.rename(columns={'how_old_are_you?': 'age'})
```

```python
# Create hierarchical index
df.set_index(['timestamp', 'age'])
```

```python
# count data entries by date
df.groupby('timestamp').describe()
```

```python
# count data entries by age
df.groupby('age').describe()
```

```python
# bin dates by early and late entries
bins = [np.datetime64(date(2015,10,22)), np.datetime64(date(2015,10,27)), np.datetime64(date(2015,10,31))]
group_names = ['early entry', 'late entry']
df['week'] = pd.cut(x=df['timestamp'], bins=bins, labels=group_names )
```
