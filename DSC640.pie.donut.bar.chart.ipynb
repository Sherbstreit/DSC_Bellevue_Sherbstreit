```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('obama-approval-ratings.xls')
df = df.set_index('Issue')
df.head()
```

```python
# transpose for pie and donut charts
df2 = df.T
df2.head()
```

```python
# set first row as new header
new_header = df2.iloc[0]
df2 = df2[1:]
df2.columns = new_header
df2.head()
```

```python
# replace spaces with underscores for easy recall
df2.columns = [c.replace(' ', '_') for c in df2.columns]
```

```python
# create pie chart
# set index names as labels
labels = df2.index
colors = ['lightskyblue', 'red', 'green']
# select variable and formatting
plt.pie(df2['Race_Relations'], labels = labels, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('Obama Approval Ratings on Race Relations', fontsize = 15) 
plt.show()
```

```python
# create donut chart
colors = ['lightskyblue', 'red', 'green']
plt.pie(df2['Race_Relations'], labels = df2.index, colors = colors)

# Add white circle to pie chart
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.title('Obama Approval Ratings on Race Relations', fontsize = 15) 

plt.show()
```

```python
# bar chart
ax = df.plot(kind='bar', figsize=(30,15), fontsize = 25)
ax.set_ylabel('Percent', fontsize = 25)
ax.set_xlabel('Issue', fontsize = 25)
patches, labels = ax.get_legend_handles_labels()
ax.legend(patches, labels, loc='best', fontsize=22)
plt.gcf().suptitle('Obama Approval Ratings by Issue',fontsize=30)
```

```python
# stacked bar chart
ax = df.plot(kind='bar', stacked=True, figsize=(30,15), fontsize = 25)
ax.set_xlabel('Issue', fontsize = 25)
patches, labels = ax.get_legend_handles_labels()
ax.legend(patches, labels, loc='best', bbox_to_anchor=(1.0, 0.5), fontsize=22)
plt.gcf().suptitle('Obama Approval Ratings by Issue',fontsize=30)
```
