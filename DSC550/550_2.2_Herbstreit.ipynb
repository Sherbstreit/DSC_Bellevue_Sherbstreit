```python
import pandas as pd
import numpy as np
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
import nltk
```

```python
data = pd.read_json('controversial-comments.jsonl', lines=True)
```

```python
data
```

```python
data['txt'] = data['txt'].str.lower()
```

```python
data['txt'] = data['txt'].str.replace('[{}]'.format(string.punctuation), '')
```

```python
data
```

```python
data['txt'] = data.apply(lambda row: nltk.word_tokenize(row['txt']), axis=1)
```

```python
stop = stopwords.words('english')
```

```python
data['txt'] = data['txt'].apply(lambda x: [item for item in x if item not in stop])
```

```python
data
```

```python
porter = PorterStemmer()
text = data['txt']
my_data_port_stemmed = []
i = 0
for i in range(len(text)):
    my_data_port_stemmed.append([porter.stem(word) for word in text[i]])
```

```python
my_data_port_stemmed
```

```python
dict_list ={ i : my_data_port_stemmed[i] for i in range(0, len(my_data_port_stemmed) ) }
dict_list
```

```python
for word in my_data_port_stemmed:
    if dict_list[word] is None:
        dict_list[word] = 1
    else:
        dict_list[word] += 1
dict_list
```

```python
sort_list = sorted(dict_list.items(), key=lambda x: x[1], reverse=True)
```
