```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#import data
df = sns.load_dataset("penguins")
```

```python
print('My most effective graph is shown below.')

# Create boxplot
ax = sns.catplot(x="species", y="bill_length_mm", hue="sex", kind="box",
                 palette=["blue", "pink"],legend=False, data=df)
# Name axis labels
ax.set(ylabel="Bill length in mm", xlabel ="Species")

# Title graph and format it
ax.fig.suptitle('Boxplot of Penguin Bill Length by Species and Sex', 
               fontsize=14, y=1.08, x=0.6)
# Move legend
plt.legend(loc='upper left')

# Remove object information text
plt.ioff()
```

```python
print('My second most effective graph is shown below.')

# Create swarm scatterplot graph
ax = sns.swarmplot(data=df, x="bill_length_mm", y="species", hue="sex",
                  palette=["blue", "pink"])

# Name axis labels
ax.set(xlabel="Bill length in mm", ylabel = "Species")

# Give title and format
ax.set_title('Scatterplot of Penguin Bill Length by Species and Sex',
            fontsize=14, y=1.08)

# Remove object information text
plt.ioff()
```

```python
print('My least effective graph is shown below.')

# Create bar graph
ax = sns.catplot(data=df, kind="bar", x="bill_length_mm",
                 y="species", hue="sex", palette=["blue", "pink"])
# Name axes
ax.set(xlabel="Bill length in mm", ylabel = "Species")

# Create title for graph
ax.fig.suptitle('Bar graph of Penguin Bill Length by Species and Sex',
                fontsize=14, y=1.06, x=0.5)

# Remove object information text
plt.ioff()
```
