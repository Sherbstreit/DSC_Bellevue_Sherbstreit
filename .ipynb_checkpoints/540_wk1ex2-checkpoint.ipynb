```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel (r'world-population.xlsm')
```

```python
Year = df['Year']
Population = df['Population']
```

```python
plt.plot(Year, Population)
plt.title('Population Vs Year')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```
