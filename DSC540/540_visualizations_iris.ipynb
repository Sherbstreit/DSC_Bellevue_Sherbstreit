```python
from sklearn import datasets
import seaborn as sns
import matplotlib as plt
import pandas as pd
```

```python
# load wine data from sklearn
iris = datasets.load_iris()
```

```python
# convert to pandas dataframe
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
df.head()
```

```python
# histogram
hist = sns.distplot( df['sepal length (cm)'], bins=10 ).set_title('Seaborn histogram of sepal length')
```

```python
# scatterplot
scatter = sns.regplot(x=df['sepal length (cm)'], y=df['sepal width (cm)']).set_title('Seaborn scatterplot of sepal length and width')
```

```python
# density graph
plot1=sns.kdeplot(df['petal length (cm)'], shade=True, color="r").set_title('Seaborn density graph of petal length and width')
plot1=sns.kdeplot(df['petal width (cm)'], shade=True, color="b")
```
