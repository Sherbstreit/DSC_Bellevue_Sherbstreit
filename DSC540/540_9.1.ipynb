```python
import urllib.request, urllib.parse, urllib.error
import requests
import re
import ssl
from bs4 import BeautifulSoup
```

```python
url = 'https://www.gutenberg.org/browse/scores/top'
```

```python
#ignore ssl errors
sslcert = ssl.create_default_context()
sslcert.check_hostname = False
sslcert.verify_mode = ssl.CERT_NONE
```

```python
# read hmtl
response = requests.get(url)
```

```python
#create code to check request status
def check_stat(request):
    if request.status_code == 200:
        print('Request successful!')
    else:
        print('Request failed.')
```

```python
check_stat(response)
```

```python
# decode html and pass onto beautful soup for parsing
contents = response.content.decode(response.encoding)
soup = BeautifulSoup(contents, 'html.parser')
```

```python
# initialize list and add all href tags
link_list = []
for link in soup.find_all('a'):
    link_list.append(link.get('href'))
```

```python
# find where ebboks start in list
link_list[:40]
```

```python
top100 = []
```

```python
# find list indices with ebooks and strip text, append to list
for i in range(33,133):
    link = link_list[i]
    link = link.strip()
    n = re.findall('[0-9]+', link)
    if len(n) == 1:
        top100.append(int(n[0]))
```

```python
top100
```

```python
# view the text
print(soup.text[:8000])
```

```python
titles = []
```

```python
# start where index is 'Top 100...'
start_idx = soup.text.splitlines().index('Top 100 EBooks yesterday')
```

```python
# append all strings after starting index
for i in range(120):
    titles.append(soup.text.splitlines()[start_idx+i])
```

```python
# select only strings containing ebooks, excluding numbers
top_titles=[]
for i in range(10, 110):
    id1,id2=re.match('^[a-zA-Z ]*',titles[i]).span()
    top_titles.append(titles[i][id1:id2])
```

```python
# view top titles
top_titles
```
