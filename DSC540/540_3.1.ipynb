```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Open file
df = pd.read_csv("Boston_housing.csv")
```

```python
# Visualize file
df.head(10)
```

```python
# number of entries in df
len(df)
```

```python
# Remove unwanted columns and save to new df
df2 = df.drop(['CHAS', 'NOX', 'B', 'LSTAT'], axis=1)
```

```python
# Last 7 entries in df
df2.tail(7)
```

```python
# Make histogram of every variable
for i in df2.columns:
    plt.hist(df2[i])
    plt.title("Histogram of variable: " +i)
    plt.show()
```

```python
# Scatterplot of crime vs. price
plt.scatter(df2['CRIM'], df2['PRICE'], alpha=0.5)
plt.title('Crime vs. Price Scatterplot')
plt.xlabel('CRIME')
plt.ylabel('PRICE')
plt.show()
```

```python
# log10 crime vs. price
plt.scatter(np.log10(df2['CRIM']), df2['PRICE'], alpha=0.5)
plt.title('Crime(log10) vs. Price Scatterplot')
plt.xlabel('CRIME(log10)')
plt.ylabel('PRICE')
plt.show()
```

```python
# define calculations
rooms = df2['RM'].mean()
age = df2['AGE'].median()
dist = df2['DIS'].mean()
lowcost = (df2['PRICE']>20).mean()
```

```python
#Display calculations
print("Mean rooms per dwelling:", rooms)
print("Median age of dwelling:", age)
print("Mean distance to 5 Boston employment centers:", dist)
print("Percentage of dwellings under $20k:", lowcost*100,"%")
```
